# Temperature prediction


```python
from numpy import loadtxt
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
```

```python
dataset = loadtxt('temperatureTrain.csv', delimiter=',')
train_x = dataset[:,0:10]
train_y = dataset[:,10]
print(train_x.shape)
print(train_y.shape)
```

```python
normalizer = preprocessing.Normalization()
normalizer.adapt(np.array(train_x))
print(normalizer.mean.numpy())
```

```python
model = tf.keras.models.Sequential([
  normalizer,
  tf.keras.layers.Dense(24,  activation='relu'),
  layers.Dropout(0.3),
  tf.keras.layers.Dense(24, activation='relu'),
  layers.Dropout(0.3),
  tf.keras.layers.Dense(24, activation='relu'),
  layers.Dropout(0.3),
  tf.keras.layers.Dense(1, activation='linear')
])
model.summary()
```

```python
#연속형이므로 (범주형일 경우 entropy loss 사용)
model.compile(loss='mean_absolute_error', optimizer=tf.keras.optimizers.Adam(0.001))
model.fit(train_x, train_y, epochs=200, batch_size=2000)
```

```python
dataset_test = loadtxt('temperatureTest.csv', delimiter=',')
test_x = dataset_test[:,0:10]
test_y = dataset_test[:,10]
print(test_x.shape)
print(test_y.shape)
```

```python
test_results = model.evaluate(test_x, test_y, verbose=1)
test_results
```
